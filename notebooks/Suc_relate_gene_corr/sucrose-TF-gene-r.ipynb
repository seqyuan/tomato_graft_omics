```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib as mpl
mpl.rcParams['pdf.fonttype']=42
mpl.rcParams['ps.fonttype']=42

import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline

import scanpy as sc
import anndata as ad


from scipy.cluster import hierarchy
from scipy.spatial import distance
from collections import defaultdict
```

```python
adata = ad.read_h5ad("../../data/CC_DD_Exp_energy_sugar.h5ad")
```

```python
df = adata.to_df()
```

```python
TFs = pd.read_excel("sucrose-TF-gene-r.xlsx", sheet_name=0, header=None)
genes = pd.read_excel("sucrose-TF-gene-r.xlsx", sheet_name=1, header=None)
```

```python
corr_df = df.loc[:, pd.concat([TFs,genes]).drop_duplicates()[0]].corr()
```

```python
out_corr_df = corr_df.loc[TFs.drop_duplicates()[0].to_list(), genes[0].to_list()]
```

```python
TFs.shape
```

```python
out_corr_df.to_csv("suc_related_TF_gene_corr.csv", sep=",")
```

```python
fig, ax = plt.subplots(1,1, figsize=(18,3.5))
sns.heatmap(out_corr_df.T, ax =ax, cmap='vlag')
plt.tight_layout()
fig.savefig('suc_related_TF_gene_corr.pdf')
```
